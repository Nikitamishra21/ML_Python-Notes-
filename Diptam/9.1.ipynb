```python
#create figure(empty canvas)
fig=plt.figure()

#add set of axes to figure
axes= fig.add_axes([0.1,0.1,0.8,0.8])
#left, bottom, width, height, (range 0 to 1)

# plot on thar set of axes
axes.plot(x,y,'b')
axes.set_xlabel('set X label')
axes.set_ylabel('set Y label')
axes.set_title('Set title')
```

```python
#create figure(empty canvas)
fig=plt.figure()

#add set of axes to figure
axes= fig.add_axes([0.1,0.1,0.8,0.8]) #main axes
axes2= fig.add_axes([0.2,0.5,0.4,0.3]) #inset axes
#left, bottom, width, height, (range 0 to 1)

# Large figure axes 1
axes.plot(x,y,'b')
axes.set_xlabel('set X label')
axes.set_ylabel('set Y label')
axes.set_title('axes 2 title')

# Insert figure axes 2
axes2.plot(x,y,'r')
axes2.set_xlabel('set X label')
axes2.set_ylabel('set Y label')
axes2.set_title('axes 2 title')

```

```python
# use similar to plt.figure() except use tuple unpacking 
fig, axes = plt.subplots()

# now use the axes object to add stuff to plot
axes.plot(x,y,'r')
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('title')
```

```python
fig, ax = plt.subplots(figsize=(12,6))

ax.plot(x, x+1, color="red", linewidth=0.25)
ax.plot(x, x+2, color="red", linewidth=0.50)
ax.plot(x, x+3, color="red", linewidth=1.00)
ax.plot(x, x+4, color="red", linewidth=1.25)

#possible linestyle option '-', '_', '-.', ':',' step'
ax.plot(x, x+5, color='green', lw=3, linestyle='-')
ax.plot(x, x+6, color='green', lw=3, ls='-.')
ax.plot(x, x+7, color='green', lw=3, ls=':')

#custom dash
line, = ax.plot(x, x+8, color='black', lw=1.50)
line.set_dashes([5, 10, 15, 10])
#format: line length,space length, ....

#possilemarker symbols:marker = '+', 'o', '*', 's'
ax.plot(x, x+9, color="blue", lw=3, ls='-', marker='+')
ax.plot(x, x+10, color="blue", lw=3, ls='--', marker='o')
ax.plot(x, x+11, color="blue", lw=3, ls='_', marker='s')
ax.plot(x, x+12, color="blue", lw=3, ls='__', marker='1')


#marker size and color
ax.plot(x, x+13, color="purple", lw=1, ls='-', marker='0', markersize=2)
ax.plot(x, x+14, color="purple", lw=1, ls='_', marker='0', markersize=4)
#ax.plot(x, x+15, color="purple", lw=1, ls='_', marker='0', markersize=8, markerfacecolor='red')
#ax.plot(x, x+16, color="purple", lw=1, ls='-', marker='0', markersize=8, markerfacecolor='yellow')
```

```python
fig,ax = plt.subplots()

ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data',0))
#set position of  spine to x=0

ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data',0))
#set position of y spine to y=0

xx=np.linspace(-0.75,1.,100)
ax.plot(xx, xx**3);
```

```python
fig,ax = plt.subplots()

ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data',0))
#set position of x spine to x=0

ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data',0))
#set position of y spine to y=0

xx=np.linspace(-.75,2,10000)
y=xx**3
z=np.fabs(xx)
a=np.intersect1d(y,z)
ax.plot(xx, y,xx,z)

ax.scatter(a**(1/3),a,color='red');
```
