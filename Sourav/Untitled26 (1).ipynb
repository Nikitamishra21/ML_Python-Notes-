```python
import numpy as np
import pandas as pd
df=pd.DataFrame({'Bengali':[35,47,60,65,70,23,56,75,77,88],
                'English':[25,37,77,60,55,67,80,53,95,85],
                'Physics':[np.nan,np.nan,np.nan,np.nan,12,57,80,90,100,78],
                'Chemistry':[np.nan,np.nan,np.nan,np.nan,57,57,80,90,100,78],
                'Math':[np.nan,np.nan,np.nan,np.nan,75,77,99,100,67,90]})
df
```

```python
df.fillna(value='IUAM')
```

```python
df['total']=df['Bengali']+df['English']+df['Physics']+df['Chemistry']+df['Math']
df
```

```python
df['percentage']=(df['Bengali']+df['English']+df['Physics']+df['Chemistry']+df['Math'])/5
df
```

```python
df
```

```python
#if (df['Bengali']+df['English'])/2>=60:
#df2=pd.DataFrame({'Bengali':[],
#                'English':[],
 #               'Physics':[],
  #              'Chemistry':[],
   #             'Math':[]})
df2=df
for i in range(10):
    if (df.iloc[i].loc['Bengali']+df.iloc[i].loc['English'])/2>60:
        df2.loc[i]=(df.iloc[i].fillna(value=df.iloc[i].mean()))
#df.fillna(value=()
df2
```

```python
df2.dropna()
```

```python
df2['remark']=''
for i in df.index:
    if (df.loc[i].loc['Bengali'] <40) | (df.loc[i].loc['English']<40) | (df.loc[i].loc['Physics'] <40) | (df.loc[i].loc['Chemistry'] <40) | (df.loc[i].loc['Math'] <40):
        df2.set_value(i,'remark','failed')
    else:
        df2.set_value(i,'remark','passed')
df2
```

```python
df2
```

```python
for i in df2.index:
    if df2.loc[i].remark=='failed':
        df2.drop(i,axis=0,inplace=True)
df2
```
