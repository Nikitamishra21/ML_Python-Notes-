```python
sns.set_style("whitegrid");
sns.FacetGrid(df,hue="species",size=4)\
    .map(plt.scatter,"petal_length","petal_width")\
    .add_legend();
plt.show()
```

```python
plt.close();
sns.set_style("whitegrid");
sns.pairplot(df,hue="species",size=3);
plt.show()
```

```python
iris_setosa=df.loc[df["species"]=="setosa"];
iris_virginica= df.loc[df["species"]=="virginica"];
iris_versicolor= df.loc[df["species"]=="versicolor"];

plt.plot(iris_setosa["petal_length"],
        np.zeros_like(iris_setosa['petal_length']),
        'o')
plt.plot(iris_virginica["petal_length"],
        np.zeros_like(iris_setosa['petal_length']),
        'o')
plt.plot(iris_versicolor["petal_length"],
        np.zeros_like(iris_setosa['petal_length']),
        'o')
plt.show()
```

```python
counts, bin_edges=np.histogram(iris_setosa['petal_length'],bins=10,density= True)
pdf= counts/(sum(counts))
print(pdf);
print(bin_edges);
cdf= np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf);
plt.plot(bin_edges[1:],cdf)
counts, bin_edges= np.histogram(iris_setosa['petal_length'],bins=20, density= True)
pdf= counts/(sum(counts))
plt.plot(bin_edges[1:],pdf);

plt.show()
```

```python
pdf - percenage of total data points at that value
cdf - percentage of data points value below X0
```
