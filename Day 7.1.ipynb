```python
left=pd.DataFrame({'key':['k0','k1','k2','k3'],
                  'B':['B4','B5','B6','B7'],
                  'C':['C4','C5','C6','C7']})

right=pd.DataFrame({'key':['k0','k1','k1','k3'],
                  'D':['D4','D5','D6','D7'],
                  'E':['E4','E5','E6','E7']})

print (left)
print (right)


#keys in right is a bit changed to get a different output in various joint
```

```python
pd.merge(left,right,how='inner', on='key')
```

```python
pd.merge(left,right,how='outer', on='key')
```

```python
pd.merge(left,right,how='left', on='key')
```

```python
pd.merge(left,right,how='right', on='key')
```

```python
left1=pd.DataFrame({'key1':['k0','k0','k1','k2'],
                   'key2':['k0','k1','k0','k1'],
                  'B':['B0','B1','B2','B3'],
                  'C':['C0','C1','C2','C3']})

right1=pd.DataFrame({'key1':['k0','k1','k1','k2'],
                     'key2':['k0','k0','k0','k0'],
                  'D':['D0','D1','D2','D3'],
                  'E':['E0','E1','E2','E3']})

pd.merge(left1,right1,on=['key1','key2'])
```

```python
left=pd.DataFrame({'B':['B4','B5','B6','B7'],
                  'C':['C4','C5','C6','C7']},
                 index=['k0','k1','k2','k3'])

right=pd.DataFrame({'D':['D4','D5','D6','D7'],
                  'E':['E4','E5','E6','E7']},
                  index=['k0','k0','k2','k3'])

left.join(right,how='outer')
```
